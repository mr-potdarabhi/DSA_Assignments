### Question-1

Given a binary tree, your task is to find subtree with maximum sum in tree.

Examples:

Input1 :       

       1

     /   \

   2      3

  / \    / \

4   5  6   7

Output1 : 28

As all the tree elements are positive, the largest subtree sum is equal to sum of all tree elements.

Input2 :

       1

     /    \

  -2      3

  / \    /  \

4   5  -6   2

Output2 : 7

Subtree with largest sum is :

 -2

 / \

4   5

Also, entire tree sum is also 7.

</aside>

```python
import math

class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

def maxSubtreeSum(root):
    if root is None:
        return 0
    
    leftSum = maxSubtreeSum(root.left)
    rightSum = maxSubtreeSum(root.right)
    
    subtreeSum = leftSum + rightSum + root.val
    
    if subtreeSum > maxSubtreeSum.maxSum:
        maxSubtreeSum.maxSum = subtreeSum
        maxSubtreeSum.subtreeRoot = root
    
    return subtreeSum

def find_Max_Sub_tree_Sum(root):
    maxSubtreeSum.maxSum = -math.inf
    maxSubtreeSum.subtreeRoot = None
    maxSubtreeSum(root)
    
    return maxSubtreeSum.maxSum

# Example 1:
# Create the binary tree from the example
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.right.left = Node(6)
root.right.right = Node(7)

maxSum = find_Max_Sub_tree_Sum(root)
print("Maximum subtree sum:", maxSum)
```

```python
# Example 2:

root2 = Node(1)
root2.left = Node(-2)
root2.right = Node(3)
root2.left.left = Node(4)
root2.left.right = Node(5)
root2.right.left = Node(-6)
root2.right.right = Node(2)

maxSum2 = find_Max_Sub_tree_Sum(root2)
print("Maximum subtree sum :", maxSum2)
```

### Question-2

Construct the BST (Binary Search Tree) from its given level order traversal.

Example:

Input: arr[] = {7, 4, 12, 3, 6, 8, 1, 5, 10}

Output: BST:

            7

         /    \

       4     12

     /  \     /

    3   6  8

   /    /     \

 1    5      10

</aside>

```python
from collections import deque
# Function to construct the BST from level order traversal
def constructBST(arr):
    if not arr:
        return None
    
    # Create the root node using the first element of the array
    root = Node(arr[0])
    queue = deque([root])  # Initialize a queue with the root node
    index = 1  # Keep track of the current index in the array
    
    while queue:
        currNode = queue.popleft()  # Dequeue a node from the queue
        
        # Create the left child of the current node
        if index < len(arr):
            leftVal = arr[index]
            if leftVal is not None:  # Check if the value is not None
                currNode.left = Node(leftVal)
                queue.append(currNode.left)  # Enqueue the left child
            index += 1
        
        # Create the right child of the current node
        if index < len(arr):
            rightVal = arr[index]
            if rightVal is not None:  # Check if the value is not None
                currNode.right = Node(rightVal)
                queue.append(currNode.right)  # Enqueue the right child
            index += 1
    
    return root  # Return the root node of the constructed BST

# Example usage:
levelOrder = [7, 4, 12, 3, 6, 8, 1, 5, 10]
root = constructBST(levelOrder)

# Helper function to print the BST in inorder traversal
def inorderTraversal(node):
    if node:
        inorderTraversal(node.left)
        print(node.val, end=' ')
        inorderTraversal(node.right)

# Print the BST in inorder traversal
print("BST:")
inorderTraversal(root)
```

### Question-3

Given an array of size n. The problem is to check whether the given array can represent the level order traversal of a Binary Search Tree or not.

Examples:

Input1 : arr[] = {7, 4, 12, 3, 6, 8, 1, 5, 10}

Output1 : Yes

For the given arr[], the Binary Search Tree is:

            7

         /    \

       4     12

     /  \     /

    3   6  8

   /    /     \

 1    5      10

Input2 : arr[] = {11, 6, 13, 5, 12, 10}

Output2 : No

The given arr[] does not represent the level order traversal of a BST.

</aside>

```python
def isLevelOrderBST(arr):
    if not arr:
        return True
    
    n = len(arr)
    stack = []  # Initialize an empty stack
    rootVal = float("-inf")  # Set root value to negative infinity
    
    for i in range(n):
        currVal = arr[i]
        
        if currVal < rootVal:
            return False
        
        # Process elements greater than the root value
        while stack and currVal > stack[-1]:
            rootVal = stack.pop()
        
        stack.append(currVal)  # Push the current element onto the stack
    
    return True  # Return True if the array is valid and all elements have been processed; False otherwise

# Example 1:
levelOrder1 = [7, 4, 12, 3, 6, 8, 1, 5, 10]
result1 = isLevelOrderBST(levelOrder1)
print("Can represent level order of BST?:", result1)

# Example 2:

levelOrder2 = [11, 6, 13, 5, 12, 10]
result2 = isLevelOrderBST(levelOrder2)
print("Can represent level order of BST?:", result2)
```
